```python
import pandas as pd 
from sklearn.linear_model import LinearRegression #Linear Model
from sklearn.tree import DecisionTreeRegressor # Import Decision Tree Classifier
from sklearn.ensemble import RandomForestRegressor #Ensemble Learning Bagging
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn.metrics import mean_squared_error as MSE #Import scikit-learn metrics module for accuracy calculation 
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn import preprocessing 
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import make_scorer, r2_score
```

```python
data=pd.read_csv(r"Pictures\house price prediction ames city.csv")
data.head()
```

### Label Encoding

```python
#make an instance of Label Encoder
label_encoder = preprocessing.LabelEncoder()
data['MSZoning'] = label_encoder.fit_transform(data['MSZoning'])
data['Neighborhood'] = label_encoder.fit_transform(data['Neighborhood'])
data['BsmtQual'] = label_encoder.fit_transform(data['BsmtQual'])
data['HeatingQC'] = label_encoder.fit_transform(data['HeatingQC'])
data['CentralAir'] = label_encoder.fit_transform(data['CentralAir'])
data['KitchenQual'] = label_encoder.fit_transform(data['KitchenQual'])
data['GarageType'] = label_encoder.fit_transform(data['GarageType'])
data['GarageFinish'] = label_encoder.fit_transform(data['GarageFinish'])
data['PavedDrive'] = label_encoder.fit_transform(data['PavedDrive'])
data['SaleCondition'] = label_encoder.fit_transform(data['SaleCondition'])
data.drop('Id', inplace=True, axis=1)
data.head()
```

```python
X = data.drop('SalePrice', axis=1)
y = data.SalePrice
print('Shape of X = ',X.shape)
print('Shape of y = ',y.shape)
```

```python
X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.3, random_state=1)
```

### Scaling the variables with normalization

```python
mmc = MinMaxScaler()
mmc.fit(X_train)
X_train=mmc.transform(X_train)
X_test=mmc.transform(X_test)
```

### Linear Regression

```python
# Train and test Linear Regression 
regrlr = LinearRegression()
regrlr.fit(X_train, y_train)
#Predict the response for test dataset
y_pred = regrlr.predict(X_test)
r_lr= r2_score(y_test,y_pred)
print('R2 Score of Linear Regression')
print(r_lr)
```

### Decision Tree

```python
# Train and test Decision Tree 
clf = DecisionTreeRegressor(random_state = 0) 
clf = clf.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf.predict(X_test)
r_dt= r2_score(y_test,y_pred)
print('R2 Score of Decision Tree')
print(r_dt)
```

### Random Forest

```python
# Train and test Random Forest 
clf_rf = RandomForestRegressor(random_state=0)
clf_rf = clf_rf.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf_rf.predict(X_test)
r_rf= r2_score(y_test,y_pred)
print('R2 Score of Random Forest')
print(r_rf)
```

### Extreme Gradient Boosting

```python
# Train and test Extreme Gradient Boosting
clf_xgb = XGBRegressor()
clf_xgb = clf_xgb.fit(X_train,y_train)
#Predict the response for test dataset
y_pred = clf_xgb.predict(X_test)
r_xgb= r2_score(y_test,y_pred)
print('R2 Score of Extreme Gradient Boosting')
print(r_xgb)
```
